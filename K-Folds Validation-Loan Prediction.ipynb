# Loan Prediction

## Introduction

This data represents customer records of Dream Housing Finance Company.The organization deals in all home loans. They cover across all urban, semi urban and rural areas. In order to understand the workflow of the company,Customers first apply for home loan. After the application has been submitted,the company validates the customer eligibility for loan. Dream Housing wants to automate the loan eligibility process (real time) based on customer detail provided while filling online application form. These details are Gender, Marital Status, Education, Number of Dependents, Income, Loan Amount, Credit History and others. To automate this process, they have given a problem to identify the customers segments, those are eligible for loan amount so that they can specifically target these customers.

## Problem Statement

As per the company requirement to grant loan or not. It is derived as a binary classification problem which is often faced by plethora of retail banks all across the world. If the problem has been identified successfully, automation of this task would certainly save a lot of man hours to complete the task.

## Hypothesis

According to the given problem statement, we can assume the possible factors that can affect the outcome.According to the given columns of records in the dataset, we can list some factors that can affect the the loan approval by the finance company.They are as follows:
### Salary
###### Applicants and Co-Applicants with higher salary are likely to receive the loan
### Income
###### The total income of applicant's and coapplicant is likely to affect the loan status
###  Previous History
###### According to the previous loan history, it has higher chances of approval
### Loan amount
###### The approval depends on the amount of loan the customer has requested.Lesser the loan higher can be the approval
### Loan term
###### Time peroid of loan also affects and creates more chances of approval
### EMI
###### Lesser the amount to be paid monthly can also determine the approval of the loan

###### Based on the given dataset, we will try to validate each of these hypothesis using different statistical methods and machine learning approaches.

#### Salary
######  Applicants and Co-Applicants with higher salary are likely to receive the loan

```python
def gender_loan():
    gender_loan=pd.crosstab(train['Gender'],train['Loan_Status'])
    gender = gender_loan.div(gender_loan.sum(1).astype(float),axis=0)
    return gender
def married_loan():
     married_loan = pd.crosstab(train.Married,train.Loan_Status)
     married = married_loan.div(married_loan.sum(1).astype(float),axis=0)
     return married
def dependents_loan():
    dependents_loan = pd.crosstab(train.Dependents,train.Loan_Status)
    dependents = dependents_loan.div(dependents_loan.sum(1).astype(float),axis=0)
    return dependents
def education_loan():
    education_loan = pd.crosstab(train.Education,train.Loan_Status)
    education = education_loan.div(education_loan.sum(1).astype(float),axis=0)
    return education
def self_employed_loan():
    self_employed_loan = pd.crosstab(train.Self_Employed,train.Loan_Status)
    self_employed = self_employed_loan.div(self_employed_loan.sum(1).astype(float),axis=0)
    return self_employed
def credit_history_loan():
    credit_history_loan =pd.crosstab(train.Credit_History,train.Loan_Status) 
    credit_history = credit_history_loan.div(credit_history_loan.sum(1).astype(float),axis=0)
    return credit_history
def property_area_loan():
    property_area_loan = pd.crosstab(train.Property_Area,train.Loan_Status) 
    property_area = property_area_loan.div(property_area_loan.sum(1).astype(float),axis=0)
    return property_area
```

## Data Preprocessing

### Missing Values and Outliers

As we have noticed our dataset contains misings values (NAN) and extreme values in distribution which are usually treated as outliers.Our initial step would be to impute the missing values and use some functions to reduce the outliers in the distribution of data to imporve the accuracy of the model.

```python
#Remove Created Variables
train = train.drop(['Income_bin','Coapp_Income_bin','Total_Income_bin','LoanAmount_bin','Loan_Amount_Term_bin','EMI_loan_bin'],axis=1)
```

```python
#Identify the frequency of missing values
train.isna().sum()
```

### Imputation of Missing Variables

Our dataset consists of missing values in different columns like Gender,Married,Dependents,Self_Employed and Credit_History which are categorical variables.Since the missing values od these categories are less compared to the shape of the data, we can use the mode or the maximum value for imputation.Similarly, columns like LoanAmount are numerical variables which can be imputated by using the median or the mean of the data. However, in the Loan_Amount_Term we have noticed that the data contains value 360.0 with the maximum frequency which would be useful to fill in the missing values.

```python
#For Categorical variable
def impute_category():
    train['Gender'].fillna(train['Gender'].mode()[0],inplace=True)
    train['Married'].fillna(train['Married'].mode()[0],inplace=True)
    train['Dependents'].fillna(train['Dependents'].mode()[0],inplace=True)
    train['Self_Employed'].fillna(train['Self_Employed'].mode()[0],inplace=True)
    train['Credit_History'].fillna(train['Credit_History'].mode()[0],inplace=True)
    train['LoanAmount'].fillna(train['LoanAmount'].median(),inplace=True)
    train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mode()[0],inplace=True)
```

```python
#Implement Method for imputation
impute_category()
train.isna().sum()
```

We have successfully imputated the missing values using statistical methods according to the category of the variables. Applying the best technique will help us improve our model's accuracy to some extent.

### Outlier Treatment

Outliers are the extreme values in the data which affects the distribution of the variable. It is one of the main cause of variation which we saw while visualizing the loan amount. Inorder to avoid such complications we use a logarithmic transformation as it does not affect smaller values but reduces the larger values to create a normal distribution of the data.


```python
train['LoanAmount_log'] = np.log(train['LoanAmount'])
sns.distplot(train['LoanAmount_log'],color="k",bins=20)
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
```

## Predictive Modeling
### Logistic Regression
#### Training Data

Logistic regression is the appropriate regression analysis to conduct when the dependent variable is dichotomous (binary).  Like all regression analyses, the logistic regression is a predictive analysis.  It is used to describe data and to explain the relationship between one dependent binary variable and one or more nominal, ordinal, interval or ratio-level independent variables. 

```python
train = train.drop('Loan_ID',axis=1)
train.head()
```

```python
#Separate the target variable from the dataset
X = train.drop('Loan_Status',1)
y = train.Loan_Status
#Converting categorical variable into numerical variable 
X = pd.get_dummies(X)
train = pd.get_dummies(train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.3)
model = LogisticRegression(multi_class='ovr',n_jobs=1,random_state=1,solver='liblinear')
model.fit(X_train,y_train)
```

```python
predicted_val = model.predict(X_val)
score = accuracy_score(y_val,predicted_val)
score
```

###### Result: Oure prediction score is about 83% accurate which means we have predicted the targeted variable i.e. Loan_Status about 83% correctly.

To resolve this problem, we use Stratified K-fold Cross Validation technique to avoide bias and variance in our model and is a best way to work on binary classification problem.
